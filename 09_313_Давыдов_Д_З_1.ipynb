Функции и процедуры

Задача 1

изображение.png

```python
def count_dividors(n):
  return sum(1 for x in range(1, n+1) if n % x == 0)

def interval_search(m, n):
  m = -10000
  m_n = 0
  for i in range(m, n+1):
    d = count_dividors(i)
    if d > m:
      m = d
      m_n = i
  return m_n

print(interval_search(7, 20))
```

```python
def gcd(a, b):
    while b != 0:
        a, b = b, a % b
    return a

def add_fractions(A, B, C, D):
    numerator = A * D + B * C
    denominator = B * D

    common = gcd(numerator, denominator)
    num //= common
    denom //= common

    return num, denom

# Пример использования
A = int(input())
B = int(input())
C = int(input())
D = int(input())

num, den = add_fractions(A, B, C, D)
print(f"Сумма дробей: {num}/{den}")
```
