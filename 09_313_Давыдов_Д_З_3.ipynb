изображение.png

```python
n, m = map(int, input().split())
matrix = [[i * j for j in range(1, m + 1)] for i in range(1, n + 1)]

max_i, max_j, min_i, min_j = 0, 0, 0, 0

for i in range(n):
  for j in range(m):
    if matrix[i][j] > matrix[max_i][max_j]:
      max_i, max_j = i, j
    if matrix[i][j] < matrix[min_i][min_j]:
      min_i, min_j = i, j

matrix[max_i][max_j], matrix[min_i][min_j] = matrix[min_i][min_j], matrix[max_i][max_j]
for row in matrix:
  print(*row)
```

# **Словари**

изображение.png

```python
def count_words(t: str):
  a = t.split()
  res = {}

  for item in a:
    res[item] = res.get(item, 0) + 1
  return res

s = 'sfd sd sd  ew w e r sd  sdc sxj lksal wi ao '
for k, w in count_words(s).items():
  print(f'{k} - {w}')
```

изображение.png

```python
def reverse_dict(d):
  res = {}

  for k, v in d.items():
    if v in res:
      if not isinstance(res[v], list):
        res[v] = [res[v]]
      res[v].append(k)
    else:
      res[v] = k

  return res

d = {
    'cool': 1,
    'faaa': 2,
    'iiiii': 1,
    'iiiasdii': 1,
    'iiiasdasdii': 1,
}

print(reverse_dict(d))
```

```python
def min_key(d):
  return min(d, key=lambda x: d[x])

a = {
    'one': 21321,
    'two': 21221,
    'three': 1292,
    'four': 1,
    'five': 1201
}

print(min_key(a))
```

```python
def del_keys(d, lst):
  for item in lst:
    d.pop(item, None)

a = {
    'one': 21321,
    'two': 21221,
    'three': 1292,
    'four': 1,
    'five': 1201
}

del_keys(a, ['two', 'five'])
print(a)
```

```python
def find_val(d, val):
  res = []
  for k in d:
    if isinstance(d[k], list) and val in d[k] or d[k] == val:
      res.append(k)
  return res

d = {
    'cool': 1,
    'faaa': 2,
    'iiiii': 1,
    'iiiasdii': 1,
    'iiiasdasdii': 1,
    'wiwiwi': [1, 2, 3, 4, 5, 6, 7]
}

print(find_val(d, 2))
```

```python
def intersect_dicts(d1, d2):
  return {k: v for k, v in d1.items() if k in d2 and d2[k] == v}

a = {
    'one': 21321,
    'two': 21224,
    'three': 1292,
    'four': 1,
    'five': 1201
}

b = {
    'six': 21321,
    'two': 21224,
    'seven': 1292,
    'four': 1,
    'eight': 1201
}

print(intersect_dicts(a, b))
```

```python
def sup_set(a, b):
  for item in b:
    if item not in a:
      return False
  return True

a = {2, 3, 4, 5, 6, 7}
b = {5, 7}

print(['А не является надмножеством B', 'А является надмножеством B'][sup_set(a, b)])
```

изображение.png

```python
def only_elems(a, b):
  result = set()

  for elem in a:
      if elem not in b:
          result.add(elem)

  for elem in b:
      if elem not in a:
          result.add(elem)

  return result

a = {2, 3, 4, 5, 6, 7}
b = {5, 6, 7, 8, 10}

print(only_elems(a, b))
```

изображение.png

```python
def inter_set(a, b):
  res = set()

  for elem in a:
    if elem in a and elem in b:
      res.add(elem)

  return res

a = {2, 3, 4, 5, 6, 7}
b = {5, 6, 7, 8, 10}

print(inter_set(a, b))
```

изображение.png

```python
def len_unique(a):
  return len(list(set(a)))

a = [2, 1, 4, 5, 3, 23, 1, 2, 3, 23, 5, 6, 3, 2, 21, 5]

print(len_unique(a))
```

изображение.png

```python
def count_elems(a):
  res = {}
  for item in set(a):
    res[item] = a.count(item)

  return res

a = [2, 1, 4, 5, 3, 23, 1, 2, 3, 23, 5, 6, 3, 2, 21, 5]

print(count_elems(a))
```

изображение.png

```python
def union_sets(a):
  res = set()

  for st in a:
    for item in st:
      res.add(item)

  return res

a = {1, 2, 3, 4, 5}
b = {4, 5, 6, 7, 8}
c = {7, 8, 9, 10, 11}

print(union_sets([a, b, c]))
```

```python
def is_in_tuple(x, a):
  for item in a:
    if x == item:
      return True
  return False

x = 10
a = (1, 2, 3, 4, 5, 5, 3, 1, 235, 6, 3, 10)

print(is_in_tuple(x, a))
```

```python
def concat_many_tuples(a):
  res = ()

  for tp in a:
    res += tp

  return res

a = (1, 3, 4, 5, 6)
b = (12, 324, 43, 34, 11)
c = (91, 20, 30, 402, 10)

print(concat_many_tuples([a, b, c]))
```
