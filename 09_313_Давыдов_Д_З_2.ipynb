изображение.png

```python
class Gadget:
  def __init__(self, brand, model):
    self.brand = brand
    self.model = model

  def display_info(self):
    print(f'Устройство марки - {self.brand}')
    print(f'Модель - {self.model}')

class Computer(Gadget):
  def __init__(self, brand, model, age):
    super().__init__(brand, model)
    self.age = age

  def update_model(self, new_model):
    self.model = new_model

  def display_computer_info(self):
    self.display_info()
    print(f'Год выпуска - {self.age}')

a = Gadget('Xiaomi', 'Redmi 5')
b = Computer('MSI', 'Boring', 2037)
a.display_info()
print()
b.update_model('Turbo Gaming Ultra Station V4300403')
b.display_computer_info()
```

Инкапсуляция

изображение.png

```python
class Computer:
  def __init__(self, brand, model, release_year):
    self.__brand = brand
    self.__model = model
    self.__release_year = release_year

  def set_brand(self, brand):
    if brand:
      self.__brand = brand
    else:
      print('Марка не может быть пустой строкой\n')

  def set_model(self, model):
    if model:
      self.__model = model
    else:
      print('Модель не может быть пустой строкой\n')

  def set_release_year(self, release_year):
    if 1970 < release_year < 2025:
      self.__release_year = release_year
    else:
      print('Недопустмый год выпуска\n')

  @property
  def brand(self):
    return self.__brand

  @property
  def model(self):
    return self.__model

  @property
  def release_year(self):
    return self.__release_year

  def display_info(self):
    print(f'Компьютер марки - {self.__brand}')
    print(f'Модель - {self.__model}')
    print(f'Год выпуска - {self.__release_year}\n')

a = Computer("Lenovo", "Flexible", 2019)
a.display_info()
a.set_release_year(-3486)
a.set_model('Unflexible')
a.display_info()
```

Композиция

изображение.png

```python
class Restaurant:
  def __init__(self, title) -> None:
    self.title = title
    self.menu = self.Menu()

  def add_to_menu(self, dish, price):
    self.menu.add_dish(dish, price)

  def show_menu(self):
    print(f'Restaurant "{self.title}"')
    for dish, price in self.menu.dishes:
      print(f'{dish} - {price}$')

  class Menu:
    def __init__(self):
      self.dish_list = []

    def add_dish(self, dish, price):
      self.dish_list.append((dish, price))

    @property
    def dishes(self):
      return self.dish_list

r = Restaurant('Claude Monet')
r.add_to_menu('Chicken Alfredo', 8)
r.add_to_menu('Onion Soup', 10)
r.add_to_menu('Sweet and Sour Chicken', 7)
r.show_menu()
```

Агрегация

изображение.png

```python
class Firma:
  def __init__(self):
    self.projects = []

  def add_project(self, title, budget):
    self.projects.append(Project(title, budget))

  def show_projects(self):
    for i, project in enumerate(self.projects):
      print(f'{i+1}) {project}')

class Project:
  def __init__(self, title, budget):
    self.title = title
    self.budget = budget

  def __repr__(self):
    return f'Проект {self.title}, бюджет - {self.budget}$'

corp = Firma()
corp.add_project('Чат-бот', 2500)
corp.add_project('Лендинг', 1000)
corp.add_project('TODO приложение', 3000)

corp.show_projects()
```

Абстрактные классы

изображение.png

```python
import abc

class ListItem(abc.ABC):
  @abc.abstractmethod
  def get_name(self): pass

  @abc.abstractmethod
  def get_price(self): pass


class BookItem(ListItem):
  def __init__(self, name, price):
    self.name = name
    self.price = price

  @property
  def get_name(self):
    return self.name

  @property
  def get_price(self):
    return self.price


class ElectronicsItem(ListItem):
  def __init__(self, name, price):
    self.name = name
    self.price = price

  @property
  def get_name(self):
    return self.name

  @property
  def get_price(self):
    return self.price


a = BookItem('Джек Лондон', 15)
b = ElectronicsItem('Iphone 16 pro max 512gb', 1500)

print(a.get_name, "-", a.get_price)
print(b.get_name, "-", b.get_price)
```

Интерфейсы

изображение.png

```python
import abc

class IComputer(abc.ABC):
  @abc.abstractmethod
  def update_model(self, model): pass

  @abc.abstractmethod
  def get_computer_info(self): pass


class Computer(IComputer):
  def __init__(self, brand, model, year) -> None:
    self.brand = brand
    self.model = model
    self.year = year

  def update_model(self, model):
    self.model = model

  def get_computer_info(self):
    return self.brand, self.model, self.year


a = Computer('Honor', 'Magic', 2022)
print(*a.get_computer_info())
```

Полиморфизм

изображение.png

```python
class Computer:
  def __init__(self, name, model):
    self.name = name
    self.model = model

  def bootUp(self):
    raise NotImplementedError("Subclass must implement this method")

  def get_computer_info(self):
    return self.name, self.model


class WindowsComputer(Computer):
  def __init__(self, name, model):
    super().__init__(name, model)

  def bootUp(self):
    print('Запуск Windows...')


class MacComputer(Computer):
  def __init__(self, name, model):
    super().__init__(name, model)

  def bootUp(self):
    print('Запуск MacOs...')


a = WindowsComputer('MSI', 'Katana GF7000')
b = MacComputer('Macbook', 'Pro')
a.bootUp()
b.bootUp()
```
